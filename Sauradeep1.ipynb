```python
import numpy as np
import pandas as pd
df1= pd.read_csv('train_LZdllcl.csv')
df2= pd.read_csv('test_2umaH9m.csv')
```

```python
df1.head()
```

```python
#df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df1.replace(to_replace  =np.nan, value = "Bachelor's")
```

```python
df1.isnull().sum()
```

```python
df1["education"].replace(to_replace  =np.nan, value = "Bachelor's", inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1['previous_year_rating'] = df1.ffill(axis =0)
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df2["education"].replace(to_replace  =np.nan, value = "Bachelor's", inplace=True)
df2['previous_year_rating'] = df2.ffill(axis =0)
```

```python
df2.isnull().sum()
```

```python
X_train = df1.iloc[:,:-1]
X_train.head()
```

```python
Y_train = df1.iloc[:,-1]
df2.head()
```

```python
X_train.set_index('employee_id', inplace = True)
df2.set_index('employee_id', inplace = True)
X_train.head()
```

```python
one_hot_cols_train = ['department','region','education', 'gender', 'recruitment_channel']
one_hot = pd.get_dummies(X_train[one_hot_cols_train])
X_train = pd.concat([X_train,one_hot], axis = 1)
X_train.head()
```

```python
X_train.drop(['department','region','education', 'gender', 'recruitment_channel'], axis = 1, inplace= True)
X_train.head()
```

```python
one_hot_cols_train2 = ['department','region','education', 'gender', 'recruitment_channel']
one_hot2 = pd.get_dummies(df2[one_hot_cols_train2])
df2 = pd.concat([df2,one_hot2], axis = 1)
df2.drop(['department','region','education', 'gender', 'recruitment_channel'], axis = 1, inplace= True)
df2.head()
```

```python
#df2.drop(['previous_year_rating'], axis =1 , inplace= True)
#X_train.drop(['previous_year_rating'], axis =1 )
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X_train_m = pd.DataFrame(minmax.fit_transform(X_train), index = X_train.index.values, columns = X_train.columns )
df2_m = pd.DataFrame(minmax.fit_transform(df2), index = df2.index.values , columns = df2.columns )
```

```python
X_train_m.head()
```

```python
from xgboost import XGBClassifier
```

```python
modelXg1.fit(X_train_m, Y_train)
modelXg1.score (X_train_m, Y_train)
```

```python
res = modelXg1.predict(df2_m)
```

```python
result = pd.DataFrame(res, index = df2.index, columns =['is_promoted'] )
result.head()
```
